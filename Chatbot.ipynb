```python
!pip install gradio scikit-learn pandas
import pandas as pd
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import gradio as gr
```

```python

faqs = pd.read_csv('faqs.csv')
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(faqs['Question'])

def get_best_answer(user_question):
    user_vec = vectorizer.transform([user_question])
    similarities = cosine_similarity(user_vec, X)
    best_idx = similarities.argmax()
    best_score = similarities[0, best_idx]

    if best_score < 0.2:
        return "Sorry, I couldnâ€™t find a relevant answer. Could you please specify what you are asking about?"

    return faqs['Answer'].iloc[best_idx]

def chatbot_interface(user_input):
    return get_best_answer(user_input)

gr.Interface(
    fn=chatbot_interface,
    inputs=gr.Textbox(lines=2, placeholder="Ask your question..."),
    outputs=gr.Textbox(lines=4),
    title="FAQ Chatbot",
    description="Ask your question and the chatbot will match it with the closest FAQ."
).launch()
```
